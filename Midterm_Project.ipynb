```python
# 6. Remove duplicates if any
df = df.drop_duplicates()
```

## 4.5 Data Encoding

Convert categorical variables into numerical values suitable for machine learning.

```python
# 8. Encode Categorical Columns
le_gender = LabelEncoder()
df['Gender'] = le_gender.fit_transform(df['Gender'])

le_target = LabelEncoder()
df['Workout_Type'] = le_target.fit_transform(df['Workout_Type'])
```

```python
# 9. Define Features and Target
X = df.drop('Workout_Type', axis=1)
y = df['Workout_Type']
```

## 4.6 Feature Scaling

Standardize numerical variables before model training.

```python
# 10. Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## 4.7 Train-Test Split

Split the dataset into training and testing subsets.

```python
# 11. Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```
